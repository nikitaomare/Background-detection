```python
#IMPORTING 
import matplotlib.pyplot as plt
%matplotlib inline
import os
import io
import sys
import numpy as np
from sklearn.cluster import KMeans
import cv2
import tensorflow as tf
```

```python
X=[ ]

img=cv2.imread('set3_rack.jpeg')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB )
X=np.array(img)
print('X.shape :'+str(X.shape))
```

# Padding 

```python
h,w = X.shape[0],X.shape[1]
k=20
X=tf.pad(X,([0,(k-h%k)%k],[0,(k-w%k)%k],[0,0]))
print('X.Shape after padding :' + str(X.shape))
```

# Convert Tensor to NumPy

```python
sess = tf.Session()
with sess.as_default():
              Xnp=X.eval()
```

# Grid division

```python
X1=[ ]
for i in range(h//k):
    for j in range(w//k):
        grid=Xnp[i*k:i*k+k,j*k:j*k+k,:]
        X1.append(np.array(grid))
X1np=np.asarray(X1)
print('X1np.Shape :' + str(X1np.shape))
```

```python
X2np=X1np.reshape(X1np.shape[0],X1np.shape[1]*X1np.shape[2]*X1np.shape[3])
print('After reshape Xnp for Kmeans cluster:'+str(X2np.shape))
```

# Kmeans_Cluster

```python
n_clusters=5
kmeans = KMeans(n_clusters=n_clusters, random_state=0,verbose=0).fit_predict(X2np)
print('Shape of Kmeans:'+ str(kmeans.shape))
```

```python
print('Cluster Pred :'+str(kmeans[:400]))
```

# Image after Cluster division

```python
import matplotlib.pyplot as plt
%matplotlib inline
kmeans1 = kmeans.reshape(h//k, w//k)
color_map = np.zeros((h,w))

for i in range(kmeans1.shape[0]):
    for j in range(kmeans1.shape[1]):
        col=50        ##changing col  give different results
        temp =col*kmeans1[i][j] 
        color_map[i*k : i*k + k, j*k : j*k + k] = np.full((k,k),temp)
        
    
plt.imshow(color_map)
plt.show
```
